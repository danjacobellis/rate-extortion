```python
from transformers import ViTImageProcessor, ViTModel
from PIL import Image
```

```python
image = Image.open('kodak/kodim05.png'); image
```

```python
processor = ViTImageProcessor.from_pretrained('google/vit-huge-patch14-224-in21k')
model = ViTModel.from_pretrained('google/vit-huge-patch14-224-in21k')
```

```python
inputs = processor(images=image, return_tensors="pt")
outputs = model(**inputs); last_hidden_state = outputs.last_hidden_state
print(last_hidden_state.shape); print(last_hidden_state)
```
